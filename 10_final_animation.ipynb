## **1. Libraries**

# **6. Animation**

```python
## Create a list to select organ
c_list = ['Kidney','heart', 'Liver', 'Lung', 'Kidney', 'Kidney - Liver','tissue (corneas)']

##Select from the dataframe
org=d[d['organ'].isin(c_list)]
org.head()
```

```python
#org2 = org2.set_index(['year'])
d3 = org.groupby(['year','organ']).sum()
d3 = d3.reset_index()
d3.head()
```

```python
l2 = d3.pivot (index = "year",
                    columns = "organ",
                    values = "quantity")
l2 = l2.reset_index()
l2['year_str'] = l2['year'].astype(str)
l2.head()
```

```python
# The frame

fig = go.Figure(
    layout= go.Layout(
        updatemenus = [dict(type = 'buttons', direction = 'right', x = 0.9, y = 1.16)],
        xaxis = dict(range= [2015, 2022],
                     autorange = False, tickwidth = 2, dtick = 1,
                     title_text = 'year'),
        yaxis = dict(range=[20,380],
                     autorange = False,
                     title_text =''),
        title = 'Organ Donation Aases',
        title_font_size = 30,
        title_x =0.5
    )
)

## Add Traces
init = 1

## Heart
fig.add_trace(
    go.Scatter(
        x = l2.year[:init],
        y = l2.heart[:init],
        name = 'heart',
        line = dict(color='red'),
        mode = 'lines'
    )
)

## Kidney
fig.add_trace(
    go.Scatter(
        x = l2.year[:init],
        y = l2.Liver[:init],
        name = 'Liver',
        line = dict(color='blue'),
        mode = 'lines'
    )
)

## Lung
fig.add_trace(
    go.Scatter(
        x = l2.year[:init],
        y = l2.Lung[:init],
        name = 'Lung',
        line = dict(color='yellow'),
        mode = 'lines'
    )
)

## Frames
frames = [
    go.Frame(
        data = [
            go.Scatter(x=l2.year[:k], y=l2.heart[:k]),
            go.Scatter(x=l2.year[:k], y=l2.Liver[:k]),
            go.Scatter(x=l2.year[:k], y=l2.Lung[:k])
        ]
    )
    for k in range(init,len(l2)+1)
]

## Animation
fig.update(frames=frames)

## play button
fig.update_layout(
    updatemenus = [
        dict(
            buttons =list([
                dict(
                label = 'Play',
                method = 'animate',
                args = [None,{'frame':{'duration':800}}]
                )
            ]

            )
        )
    ]
)

fig.show()
```

## ***History***
In this animation, the years from 2015 to 2022 are shown on the X axis, where the number of organs donated and their quantities over the years can be observed. The heart, kidney and lung are taken because they are the most donated organs, with liver standing out among the three.
